```python
def status(value):
    if value == 9.0:
        return "Terbaik"
    elif  value == 8.0:
        return "Bagus"
    elif value == 7.0:
        return "Keren"
    else:
        return "Oke"

df['rating'] = df['home_team_fifa_rank'].map(status)
display(df.head())
```
